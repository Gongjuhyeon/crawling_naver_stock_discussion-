```python
from tqdm import tqdm
import time

from openpyxl import load_workbook
```

```python
driver = webdriver.Chrome('chromedriver',options=options)
wb2 = load_workbook(filename="/content/drive/MyDrive/Hcc lab project/dd.xlsx")
ws2 = wb2[wb2.sheetnames[3]]#1 -> sheet2 설정
print(ws2)
for i in tqdm(range(1540)):
    link=link_df['link'][i]
    url=f"https://finance.naver.com/"+str(link)
    sleep(1) 
    
    
    result1 = requests.get(url, headers = headers)
    bs = BeautifulSoup(result1.content, "html.parser")
    try:
        content=str(bs.find('div',{'class':'scr01'}).find_all(text=True))

        compi=re.compile('[^가-힣\-_]+')
        content2=re.sub(compi," ",content)
        li_content.append(content2)
        ws2.cell(row = i+2, column = 9).value =content2
    except:
        li_content.append("")

        ws2.cell(row = i+2, column = 9).value =""

wb2.save("/content/drive/MyDrive/Hcc lab project/dd.xlsx")
wb2.close()

driver.quit()
```

```python

def page(search,num,df,final_nid):
    driver = webdriver.Chrome('chromedriver',options=options)
    url=f"https://finance.naver.com/item/board.naver?code="+str(search)+"&page="+str(num)

    r = driver.get(url) #해당 url로 이동   
    
    sleep(3) 
    html1 = driver.current_url #현재 페이지 소스
    print(html1)
                               
    result = requests.get(html1, headers = headers)

   
    bs_obj = BeautifulSoup(result.content, "html.parser")
    table = bs_obj.find('table', {'class' : 'type2'})
    tt = table.select('tbody > tr')
    print("final_nid =" , final_nid)
    for i in range(2, len(tt)):
        if len(tt[i].select('td > span')) > 0:
            date = tt[i].select('td > span')[0].text
            title = tt[i].select('td.title > a')[0]['title']
            writer = tt[i].select('td.p11')[0].text.replace('\t', '').replace('\n', '')
            views = tt[i].select('td > span')[1].text
            links=tt[i].select('td.title > a') [0]['href']
            code=tt[i].select('td.title > a')[0]['href']
            code=re.split(r'[?&]',code)
            code=re.split(r'code=',code[1])
            
            nid=tt[i].select('td.title > a')[0]['href']
            nid=re.split(r'[?&]',nid)
            nid=re.split(r'nid=',nid[2])
            
            c=tt[i].find('b')
            if(c==None):
                review_count=0
                
            else:
                review_count=tt[i].select('td.title > a > span > b ')[0].text.replace('/(<[^>]+)>)/',"")
        
            sleep(2)
            
            table = pd.DataFrame({'날짜' : [date], 
                    '제목' : [title],
                    '글쓴이' : [writer],
                    '조회' : [views],
                    '댓글 수': [review_count],
                    'code': [code[1]],
                    'nid':[nid[1]],
                    'link':links
                    })
            df = df.append(table)

            if(nid[1]==final_nid):
                df =pd.DataFrame([])
            if((len(tt)-2)==i):
                final_nid=nid[1]
        
    driver.quit()
    return df,final_nid
```
